```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_excel('말뭉치_train.xlsx')
test = pd.read_excel('말뭉치_test.xlsx')
```

```python
test[:5]
```

```python
print(len(train))
print(len(test))
```

```python
data = pd.concat([train, test])
removed_data = data.drop(['번호', '연령', '성별', '상황키워드', '신체질환', '시스템응답1', '시스템응답2', '사람문장3', '시스템응답3', '사람문장4', '시스템응답4'], axis=1)
removed_data.index = [i for i in range(len(removed_data))]
```

```python
removed_data.isnull().sum()
```

```python
for i in range(1, 3):
    removed_data[f'사람문장{i}'] = removed_data[f'사람문장{i}'].fillna('')
```

```python
removed_data[:5]
```

```python
sentence_list = []
for i in range(len(removed_data)):
    sentence_list.append(f"{removed_data.iloc[i]['사람문장1']} {removed_data.iloc[i]['사람문장2']}".strip())
```

```python
sentence_list[:5]
```

```python
removed_data['문장'] = sentence_list
```

```python
removed_data
```

```python
removed_data = removed_data.drop(['사람문장1', '사람문장2'], axis=1)
```

```python
removed_data
```

```python
removed_data.drop_duplicates(subset=['문장'], inplace=True) # 중복 데이터 제거
removed_data['문장'] = removed_data['문장'].str.replace('[^0-9ㄱ-ㅎㅏ-ㅣ가-힣.,!? ]','') 

removed_data['문장'] = removed_data['문장'].str.replace('^ +','') # white space 데이터를 제거
removed_data['문장'].replace('', np.nan, inplace=True) # 빈 값을 null로 설정

removed_data = removed_data.dropna(how='any') # null 데이터 제거
```

```python
removed_data
```

```python
removed_data['감정_대분류'].value_counts()
```

```python
senti1 = ['불안 ', '기쁨 ']
senti2 = ['불안', '기쁨']

removed_data['감정_대분류'] = removed_data['감정_대분류'].replace(senti1, senti2)
```

```python
removed_data['감정_대분류'].value_counts()
```

```python
removed_data['감정_소분류'].value_counts()
```

```python
removed_data.to_csv('말뭉치_전처리_짧은문장.csv', index=False, encoding='cp949')
```
